##**Deep learning.**
> - Es una tecnica de Machine learning que consiste en el entrenamiento de algoritmos matemáticos que enseñan a tomar desiciones mediante clasificación de entradas y salidas que forman una neurona artificial. Mediante estas técnicas una computadora puede asimiliar conocimiento humano.

##**Funciones de activación.**
- Forma parte del entrenamiento de una neurona artificial según caracteristica:

> - **Sigmoid**: Función binaria de clasificación (Si/No) cuya apróximación puede verse reducida al intervalo [-5, 5] (Fuera de ese dominio su aproximación puede llegar a ser 0). **Dominio: Reales | Rango [0,1].** $ f(x) = \frac{1}{1 + e^{-x}} $

> - **Tanh**: Tangente hiperbólica se define como el cociente entre el seno hiperbólico y el coseno hiperbólico del número real $ x $. Su rango se comprende entre [-1, 1] y su dominio son los reales. $ f(x) = \frac{sinh(x)}{cosh(x)} = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $

> - **ReLU (Rectificación lineal unitaria):** Función segmentada.
>> $ f(x) = \left\{ \begin{array}{lcc}
      x, & x \geq 0 \\
      \\ 0, & x < 0
      \end{array}
   \right. $

> - **Softsign:**  Función similar a tanh con la diferencia que posee un mayor dominio en el eje x. 
$ f(x) = \frac{ x }{ \left\lvert x \right\rvert + 1 } $

> - **Softplus:**  
$ f(x) = ln({e^{x}} + 1) $



##**Ejemplo: Función celsius a fahrenheit.**

```
celcius = ['32.5', '40.7', '23.0', '44.0', '15.9', '33.0', '45.6', '67.0', '70.4']
fahrenheit = []
```

```
def c2f (c):
  return (c*(9/5)) + 32
```

```
for cel in celcius:
  fahrenheit.append(round(c2f(float(cel)), 2))

print(fahrenheit)
```

##**Aplicando Deep Learning**


```
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

celcius_2 = np.array([32.5, 40.7, 23.0, 44.0, 15.9, 33.0, 45.6, 67.0, 70.4], dtype=float)
fahrenheit_2 = np.array([90.5, 105.26, 73.4, 111.2, 60.62, 91.4, 114.08, 152.6, 158.72], dtype=float)
```

```
# Creando el modelo
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```
# Compilar
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1), loss='mean_squared_error')
```

```
history = model.fit(celcius_2, fahrenheit_2, epochs=500)
```

```
print(model.predict([0]))
```

```
plt.plot(history.history['loss'])
plt.title('Evolución de la perdida durante el entrenamiento')
plt.xlabel('Epochs')
plt.ylabel('Training loss')
plt.legend('Training loss')
plt.show()
```
