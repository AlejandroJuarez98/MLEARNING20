##**Matplotlib**

> Matplotlib es una librería muy utilizada en python para generar gráficos que estan contenidos en listas y arreglo de datos.

```
import matplotlib.pyplot as plt

listy = [2,4,6,8,10,12,14]

plt.plot(listy, 'ro')
plt.ylabel('eje y')
plt.show()

list_x = [2,4,6,8,10]
list_y = [1,-3,5,-7,9]

plt.plot(list_x, list_y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')

# Rangos de los ejes
plt.axis([2, 10, -8, 10])
plt.show()
```

```
import pandas as pd

dframe = pd.read_csv('https://raw.githubusercontent.com/AlejandroJuarez98/MLEARNING20/master/tables.csv', index_col=0)
dframe.columns = dframe.columns.str.lower()
dframe.columns = dframe.columns.str.replace(' ', '_')

dframe

plt.figure(figsize=(5,5))
plt.bar(dframe['header_2'].unique(), dframe['header_2'].value_counts())

plt.ylabel("Total")
plt.xlabel("Tipos")

plt.title('Distribución de tipos')
plt.show()
```
