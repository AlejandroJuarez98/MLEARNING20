##**Pandas**
> Pandas es una librería muy utilizada para la manipulación de datos, esta librería presenta información tabulada y es una alternativa al uso de Excel.

```
# Importando la librería
import pandas as pd

dframe = pd.read_csv('https://raw.githubusercontent.com/AlejandroJuarez98/MLEARNING20/master/tables.csv', index_col=0)
dframe.head()
```

```
dframe.tail() # Últimos 5 valores
```

```
name_of_columns = dframe.columns
print(list(name_of_columns))

dframe.columns = dframe.columns.str.lower()
dframe.columns = dframe.columns.str.replace(' ', '_')

first_column_values = dframe['header_1'].values.tolist()
print(first_column_values, '\n')

dframe.head(10)
```

```
 # Reemplazar todos los valores que están antes de col 1 de la columna header_1 por ''
dframe['header_1'] = dframe['header_1'].str.replace('.*(?=col 1)', '')
dframe.head(10)
```

```
# Establecer la columna indice de la tabla
#dframe = dframe.set_index('header_1')
dframe.head()
```

```
# Esta función se utiliza para encontrar valores nulos dentro de la tabla
dframe.isnull().any()

# Detecta dentro del frame cuantos valores NaN hay y los suma
dframe.isna().sum()
```

```
# Eliminando columna de dataframe.
dframe.drop('header_8', inplace=True, axis=1)
dframe.head()
```

```
#Eliminando fila del dataframe
dframe.drop(5, inplace=True, axis=0)
dframe.head()
```

```
# Retornando valores de cada tipo de dato almacenado en una columna
dframe['header_1'].value_counts()
# Para ver mas descripciones de nuestro dataframe utilizamos
dframe.describe()
```
